```python
children = ["Andrew","Laura","Joe"]
lol_children = [[c, len(c)]  for c in children]
lol_children #a list of lists[name, length of name]
```

```python
#DataFrame from LOL
import pandas as df 
df_lol_children = df.DataFrame(lol_children,columns = ["child","charlen"])
#columns argument is optional. df_lol_children = df.DataFrame(lol_children) gives same result but columns are 0, 1
df_lol_children
```

```python

len_char = [len(c) for c in children]
#a dictionary with lists as values
d_children = {"child": children, "charlen": len_char}
d_children
```

```python
#DataFrame from dictionary
df_d_children = df.DataFrame(d_children) #keys become column names
df_d_children
```

## Get the data

```python
import pandas as pd
poll_url = 'http://projects.fivethirtyeight.com/general-model/president_general_polls_2016.csv'
#pandas has method to read csv into DataFrame. Lots of optional arguments in documentation https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html
polls = pd.read_csv(poll_url)
polls
```

```python
polls.info() #.info method of pandas DataFrame prints column names, count of non-null values, data type
```

```python
polls.columns #.columns method returns a pandas Index object containing axis labels
```

## Sort the data

```python
polls.sort_values('enddate', ascending=False, inplace=True) #.sort_values method. First argument is by, 
#which is name or list of names to sort by. Default axis is 0, i.e. sorting rows. Defaults to ascending. In place?
polls.head() #.head returns first n rows. Defaults to 5
```

## Use lists, slices, tuples, and dictionary objects

### A list

```python
# columns=['cycle','branch','matchup','forecastdate'], inplace=True
dict_input = {"columns":['cycle','branch','matchup','forecastdate'], "inplace":True}
polls.drop(**dict_input) #.drop method removes rows or columns (default) by specifying label names. **dict_input provides 
#"columns" and "inplace" keyword arguments from dictionary
polls.head(2)
```

### A dictionary, ranges and comprehension

```python
polls = polls.rename(columns={'adjpoll_clinton':'Clinton',
                      'adjpoll_trump':'Trump'}) #.rename renames columns or index labels IAW dictionary
polls.info()                      
```

```python
len(polls)
```

```python
num = [f for f in range(0,12624)]
num[0:100:10] #slice num from 0 to 100 by 10, returns list
```

### Two slices

```python
foo = polls.loc[0:100:10,'state':'grade'].index
#loc[row, column], row is not the same as index. Indexes are preserved through .sort method
",".join([str(f) for f in foo])
```
