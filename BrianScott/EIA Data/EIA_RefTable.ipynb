### Balancing Authority Dictionary

```python
from openpyxl import load_workbook
import pandas as pd

#open Excel data description file
data_ref_fname = 'EIA930_Reference_Tables.xlsx'

#read Balancing Authority table to dataframe
BAs_df = pd.read_excel(data_ref_fname, sheet_name = 'BAs')
BAs_df.head()
```

```python
#BAs_df.set_index('BA Code', verify_integrity=True, inplace=True)
BAs_df.columns = [c.replace(' ', '_') for c in BAs_df.columns]
BAs_df.columns = [c.replace('/', '_') for c in BAs_df.columns]
BAs_df.columns = [c.replace('.', '') for c in BAs_df.columns]
BAs_df.drop(BAs_df.query('US_BA == "No"').index, inplace = True)
BAs_df.head()
```

```python
BAs_trans = BAs_df.transpose()
BAs_trans.head()
```

```python
#convert BA dataframe to dict and save as json
import json

BAs_trans = BAs_trans.fillna('')
BA_dict = BAs_trans.to_dict()
print(BA_dict)
BA_dict_fname = 'BA_dict.json'
with open(BA_dict_fname, "w") as outfile:
    json.dump(BA_dict, outfile, indent=4)
```

### CO2 Emissions Table

```python
from openpyxl import load_workbook
import pandas as pd

#open Excel CO2 emissions factors table
co2_coeffs_fname = 'co2_factors.xlsx'

#read CO2 coefficients table to dataframe
co2_df = pd.read_excel(co2_coeffs_fname, sheet_name='co2factors')
co2_df
```

```python
co2_df = co2_df[['EIA_Fuel_Code', 'Fuel', 'kg_co2_per_million_btu']].set_index('EIA_Fuel_Code')
co2_df.head()
```

```python
co2_df_trans = co2_df.transpose()
```

```python
co2_df_trans.head()
```

```python
energy_source_dict = co2_df_trans.to_dict()
print(energy_source_dict)
```

```python
gencap_esrc_dict_fname = 'e_src_dict.json'
with open(gencap_esrc_dict_fname, "r") as infile:
    esrc = pd.read_json(infile).transpose()
esrc=esrc.rename(columns={'energy_source_desc':'Fuel'}).drop(index='Operating')
esrc
```

### Full CO2/Mwh by energy source dictionary

```python
esrc_full = pd.concat([co2_df, esrc])
esrc_co2 = esrc_full[~esrc_full.index.duplicated(keep='first')].fillna(0.0) \
    .sort_values(by='kg_co2_per_million_btu', ascending=False).rename(columns={'Fuel':'energy_src'})
esrc_co2
```

```python
esrc_co2_dict = esrc_co2.transpose().to_dict()
esrc_co2_dict_fname = 'esrc_co2_dict.json'
with open(esrc_co2_dict_fname, "w") as outfile:
    json.dump(esrc_co2_dict, outfile, indent=4)
```
