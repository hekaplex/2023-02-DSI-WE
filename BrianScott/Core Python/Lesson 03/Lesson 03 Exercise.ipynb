```python
import csv

def readFileToList(filereader):
    lines = []
    for row in filereader:
        lines.append(row)
    return(lines) 
```

```python
def prettyPrintSong(song):
    songInfo = {'title':song[0]['track_title'], 'artist':song[0]['artist'], 'album':song[0]['album']}
    lyrics = [line['lyric'] for line in song]
    infoString = f"{songInfo['title']}\n{songInfo['artist']}\n{songInfo['album']}\n"
    lyricString = ''
    for line in lyrics:
        lyricString = f"{lyricString}{line}\n"
    songString = f"{infoString}\n{lyricString}"
    return songString
    
```

```python
datafileName = "taylor_swift_lyrics.csv"
datafile = open(datafileName, 'r', encoding='utf-8')
reader = csv.reader(datafile)
#create list of field names from first line of data file; convert to tuple
fields = next(reader)
fields.append('line_id')
print(f"Fields:  {fields}")
allLines = readFileToList(reader)
datafile.close()
id = 1
for line in allLines:
    line.append(id)
    id += 1
print(allLines[0])
```

```python
indexDict = {k:fields.index(k) for k in fields}
print(indexDict)
```

```python
albumSet = {line[indexDict['album']] for line in allLines}
print(albumSet)
```

```python
trackSet = {line[indexDict['track_title']] for line in allLines}
print(trackSet)
```

```python

lineDicts = []
for line in allLines:
    lineDicts.append(line[indexDict['line_id']])
    lineDicts[line[indexDict['line_id']]-1]={field:line[indexDict[field]] for field in fields}
print(lineDicts[0])
```

```python
tracks = {track:[] for track in trackSet}
for line in lineDicts:
    for track in tracks:
        if track == line['track_title']:
            tracks[track].append(line)

print(tracks['Cold as You'])
```

```python
albums = {album:[] for album in albumSet}
for track in tracks:
    for album in albums:
        if album == tracks[track][0]['album']:
            albums[album].append(tracks[track])

print(albums['1989'])
```

```python
def main():
    while True:
        songtitle = input("Enter a Taylor Swift song title from 2019 or earlier, or \"q\" to exit:  ")
        if songtitle == 'q':
            break
        else:
            try:
                print(prettyPrintSong(tracks[songtitle]))
            except KeyError:
                print(f"{songtitle} is not in the data set. Please check your spelling and try again")
```

```python
main()
```
