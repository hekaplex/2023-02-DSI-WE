```python
import numpy as np
import pandas as pd
import re
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('stopwords')
```

```python
corpus = pd.read_csv('train.csv')
corpus
```

```python
corpus.columns
```

```python
corpus['comment_text'] = corpus['comment_text'].str.lower()
```

```python
corpus['comment_text']
```

```python
import re
corpus['comment_text'] = corpus['comment_text'].replace(r"(@\[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?)", "")
corpus.iloc[3,1]
```

```python
tokenizer = RegexpTokenizer(r'\w+')
```

```python
text= corpus.copy(deep= True)
text['comment_text'] = text['comment_text'].apply(lambda x:tokenizer.tokenize(x))
text
```

```python
def rm_SW(words):
    info = [w.lower() for w in words if w not in stopwords.words('english')]
    return info
```

```python
text['comment_text'] = text['comment_text'].apply(lambda x: rm_SW(x))
text
```

```python
def list_join(words):
    palabra= " ".join(w for w in words) 
    return palabra
```

```python
text['comment_text'] = text['comment_text'].apply(lambda x: list_join(x))
```
