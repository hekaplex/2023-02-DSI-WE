```python
# dropping for the subplot - boxplots below
df02 = df01.drop(columns=["Year_Obv", "States" ])
df02 = df02.rename(columns={'DB_Values':'Diab', 'OB_Values':'Obes','HC_Values':'HgCho', 'HD_Values':'HtDis', 'HBP_Values':'HbpPrs'})
```

```python
#Subplot

red_circle = dict(markerfacecolor='red', marker='o', markeredgecolor='white')
fig, axs = plt.subplots(1, len(df02.columns), figsize=(20,10))

for i, ax in enumerate(axs.flat):
    ax.boxplot(df02.iloc[:,i], flierprops=red_circle) # exclude row#0 and #1
    ax.set_title(df02.columns[i], fontsize=20, fontweight='bold')
    ax.tick_params(axis='y', labelsize=14)
    
plt.show()
```
