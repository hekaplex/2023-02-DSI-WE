```python
df.rename(columns={"Scored Labels":"predicted","price":"actual"},inplace=True)
```

```python
df["residual"] = df["actual"] - df["predicted"]
```

```python
#MAE Mean absolute error

df["abs_err"] = abs(df.residual)
```

```python
#RMSE = Root mEan Squared error
df["sqr_err"] = df.residual**2
df["sqr_mean_err"] = (df.actual - df.actual.mean())**2
df["naive_mean_err"] = df.sqr_mean_err**0.5
```
